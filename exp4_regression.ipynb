# Experiment 4: Linear and Polynomial Regression

## Objective:
To implement Linear and Polynomial Regression models and compare their performance.

## Steps:
1. Generate dataset
2. Apply Linear Regression
3. Apply Polynomial Regression
4. Compare results

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
X = np.arange(1, 10).reshape(-1, 1)
y = X**2 + 2*X + 3

plt.scatter(X, y)
plt.title("Original Data")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(X, y)

y_pred_lin = lin_model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred_lin)
plt.title("Linear Regression")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)

y_pred_poly = poly_model.predict(X_poly)

plt.scatter(X, y)
plt.plot(X, y_pred_poly)
plt.title("Polynomial Regression")
plt.show()
```

## Conclusion:
Polynomial Regression gives better results when the data is non-linear,
while Linear Regression works well for straight-line relationships.
